```python
from ultralytics import YOLO, SAM
from IPython.display import Image, display
import glob

# Load the YOLO model
yolo_model = YOLO("runs/detect/train9/weights/best.pt")  

# Run inference on an image
results = yolo_model("test_images/meningioma_3.jpg") 

# Load the SAM model
sam_model = SAM("sam2_b.pt")  

# Loop through results and apply SAM segmentation
for result in results:
    class_ids = result.boxes.cls.int().tolist() 
    if len(class_ids):
        boxes = result.boxes.xyxy  # Get bounding boxes
        sam_results = sam_model(result.orig_img, bboxes=boxes, verbose=False, save=True, device='cpu')
```
